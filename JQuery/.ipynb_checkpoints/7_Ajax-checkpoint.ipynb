## AJAX

It was developed by google when they were working with their gmail interface.

It is useful when we are developing **web apps**

<span style="background-color:yellow;"> It allows us to **grab** contents **from remote server** and notifies us **without having to refresh the page**</span>

**Downside :** we cant test it on local files. So we ll be testing it in out [hosted website](http://aravindsankaran-com.stackstaging.com/JQuery/html/7_ajax.html)

### Steps :

>1) Log in to [eco-webhosting](https://www.ecowebhosting.co.uk/cp/login)

>2) Go to **manage hostings** and open your remote folder for your webpage

>3) Unlock FTP

>4) Open **FileZilla** and go to **site manager** (top-left icon) and connect to your site 
<img src="filezilla" width="500" height="500" />

>5) Open your remote folder in **FileZilla** and start editing your file (right click -> view/edit)

>6) After every save, **FileZilla** will prompt you to relond content onto remote server.

>7) Once reloaded, the changes can be seen on your live website

### Illustration of $.ajax

```html

<html>
  <head>
    <title> AJAX </title>

    <script src="jquery-3.3.1.min.js"></script>

    <style type="text/css">

      .shape{
        height: 100px;
        width: 100px;
        margin : 30px;

      }

      #circle{
        background-color: blue;
        border-radius: 50%;
      }


    </style>

  </head>
  <body>


      <div id="circle" class="shape">

      </div>

      <p id="text">

      </p>

    <script type="text/javascript">

      //get the info from a file
      //"data" contains the text from info.txt
      $.get("info.txt", function(data){

        $("#text").html(data);

      }); // Here we cant do anything if the get function fails

      //so we go for ajax
      $.ajax("info.txt").done( function(data){

          alert("got info from AJAX " + data);
      });

      $.ajax("info1.txt")
      .done( function(data){

          alert("got info from AJAX " + data);
      })
      .fail(function(){

          alert("could not get 'info1.txt'");
      });




    </script>

  </body>

</html>
```
