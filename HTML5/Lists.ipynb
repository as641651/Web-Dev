## Ordered and unordered lists

```html

<html>
  <head>
    <title> My WebPage </title>
  </head>
  <body>

    <!-- Unordered list-->
    <h3>Unordedred list</h3>
    <ul>
      <li><strong>Tensor Flow</strong></li>
      <li><em>Torch</em></li>
      <li><ins>Theano</ins></li>
    </ul>

    <h3>Ordered list</h3>
    <!-- Ordered list -->
    <ol>
      <li><strong>Tensor Flow</strong></li>
      <li><em>Torch</em></li>
      <li><ins>Theano</ins></li>
    </ol>

    <h4>Can specify start number</h4>
    <ol start="10">
      <li><strong>Tensor Flow</strong></li>
      <li><em>Torch</em></li>
      <li><ins>Theano</ins></li>
    </ol>

    <h4>Can reverse</h4>
    <ol reversed>
      <li><strong>Tensor Flow</strong></li>
      <li><em>Torch</em></li>
      <li><ins>Theano</ins></li>
    </ol>

    <h4>Can specify type. Available types : 1,a,A,i,I</h4>
    <ol type="I"> <!-- available types1,a,A,i,I -->
      <li><strong>Tensor Flow</strong></li>
      <li><em>Torch</em></li>
      <li><ins>Theano</ins></li>
    </ol>

  </body>
</html>

```
