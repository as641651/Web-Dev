## Structure of a Web Page

```html
<!doctype html>
<html>
<head>
    <!-- This appears in the browser tab. Does not appear in the page -->
    <title>Example Domain</title> 

    <!-- Standard character set that we ll be using -->
    <meta charset="utf-8" /> 
    <meta http-equiv="Content-type" content="text/html; charset=utf-8" />
    
    <!-- "Viewport" meta tag defines the "Width of the page, which is set to "device-width" and scaling -->
    <meta name="viewport" content="width=device-width, initial-scale=1" /> 
    
    <!-- This part will be elaborated in next section -->
    <style type="text/css"> 
    body {
        background-color: #f0f0f2;
        margin: 0;
        padding: 0;
        font-family: "Open Sans", "Helvetica Neue", Helvetica, Arial, sans-serif;
        
    }
    div {
        width: 600px;
        margin: 5em auto;
        padding: 50px;
        background-color: #fff;
        border-radius: 1em;
    }
    a:link, a:visited {
        color: #38488f;
        text-decoration: none;
    }
    @media (max-width: 700px) {
        body {
            background-color: #fff;
        }
        div {
            width: auto;
            margin: 0 auto;
            border-radius: 0;
            padding: 1em;
        }
    }
    </style>    
</head>

<body>
<!-- Section inside the body -->
<div>
    <!-- Font size h1 -->
    <h1>Example Domain</h1>
    
    <!-- paragraph -->
    <p>This domain is established to be used for illustrative examples in documents. You may use this
    domain in examples without prior coordination or asking for permission.</p>
    
    <!-- "a" is for links -->
    <p><a href="http://www.iana.org/domains/example">More information...</a></p>
</div>
</body>
</html>
```
