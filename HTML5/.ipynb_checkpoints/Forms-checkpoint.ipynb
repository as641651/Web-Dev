## Forms

```html

<html>
  <head>
    <title> My WebPage </title>
  </head>
  <body>

    <!-- "Action" has the url to which form will be submitted when you click submit input -->
    <form action="https://space-facts.com/earth/">
      <!-- box where we can type something  -->
      <!-- "value" field is the default value -->
      <p>Username with default: <input type="text" value="your uname"></p>
      <!-- default value should be deleted before typing new word. So placeholders are better -->
      <p>Username with placeholder: <input type="text" placeholder="your uname"></p>

      <!-- checkbox -->
      <p>Stay logged in: <input type="checkbox"></p>
      <!-- "checked" attribute checks box by default -->
      <p>Stay logged in (checked): <input type="checkbox" checked></p>

      <!-- radio "name" tag is used to link radio buttons so that only one in that class can be checked -->
      <!-- here, with the "value" option the code can get a handle to what has been selected -->
      <p>Under 18: <input type="radio" name="age" value="u18"></p>
      <p>Over 18: <input type="radio" name = "age" value="o18"></p>

      <!-- Pretty much every form element should have a name.
      We ll see this later -->

      <!-- Dropdown box -->
      <p> Favourite food:
        <select>
          <option> Pizza</option>
          <option selected> Ice cream</option> <!-- Default selection -->
          <option> Sandwich</option>

        </select>
      </p>

      <!-- submit button-->
      <p>
        <input type="submit" value="Click me!">
      </p>


    </form>

  </body>
</html>

```
