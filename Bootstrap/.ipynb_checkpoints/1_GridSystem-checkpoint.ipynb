## Grid system

> Endables dividing the width of the page into 12 column system

> The **pre-defined** classes - **container, row, column ** are used for this purpose

```html

<div class="container">
      <!-- Creates a row -->
      <div class="row">
        <!-- Creates a column -->
        <div class="col"></div>
        <div class="col"></div>    
    </div>
</div>
```

> Automatically redefines layout depending on screen width and also on resizing. The break points to change layout can also be defined

```html
<!-- Creates a column that occupies 6/12 columns -->
<!-- "sm" indicates that the columns would stack upon each other at 576px-->
<div class="col-sm-6"></div>
```

> To know more about grid options, click [here](https://getbootstrap.com/docs/4.0/layout/grid/)

### Script

```html

<!doctype html>
<html lang="en">
  <head>
    <!-- Required meta tags -->
    <meta charset="utf-8">
    <meta name="viewport" content="width=device-width, initial-scale=1, shrink-to-fit=no">

    <!-- Bootstrap CSS -->
    <link rel="stylesheet" href="https://maxcdn.bootstrapcdn.com/bootstrap/4.0.0/css/bootstrap.min.css" integrity="sha384-Gn5384xqQ1aoWXA+058RXPxPg6fy4IWvTNh0E263XmFcJlSAwiGgFAW/dAiS6JXm" crossorigin="anonymous">

    <title>Bootstrap</title>
    <style type="text/css">

    .bg1{
      /* Notice : I'm using "background" and not "background-color" for gradients to apply */
      background: linear-gradient(to right,white,#92a9ce);
      height: 100px;

    }

    .bg2{
      background: linear-gradient(to right,white, #e2da9c);
      height: 100px;

    }

    .bg3{
      background: linear-gradient(to right,white, #93dd77);
      height : 200px;

    }

    </style>
  </head>
  <body>


    <div class="container">
      <h1>Break into three (Auto)</h1>

      <!-- Creates a row container with three columns -->

      <!-- row -->
      <div class="row">
        <!-- column with "small" breakpoint, which stacks up at 576px -->
        <div class="col-sm bg1">
          One of three columns
        </div>
        <div class="col-sm bg2">
          One of three columns
        </div>
        <div class="col-sm bg3">
          One of three columns
        </div>
      </div>

    </div>

    <div class="container">
      <h1>Break into two (Manual)</h1>

      <!-- There are totally 12 columns in a grid -->
      <!-- To break the page into 50-50,
      we create a row with two columns with 6 units each -->
      <div class="row">
        <!-- "md" stacks up at 720px -->
        <div class="col-md-6 bg1">
          50% width
        </div>
        <div class="col-md-6 bg2">
          50% width
        </div>
      </div>

    </div>



    <!-- Optional JavaScript -->
    <!-- jQuery first, then Popper.js, then Bootstrap JS -->
    <script src="https://code.jquery.com/jquery-3.2.1.slim.min.js" integrity="sha384-KJ3o2DKtIkvYIK3UENzmM7KCkRr/rE9/Qpg6aAZGJwFDMVNA/GpGFF93hXpG5KkN" crossorigin="anonymous"></script>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/popper.js/1.12.9/umd/popper.min.js" integrity="sha384-ApNbgh9B+Y1QKtv3Rn7W3mgPxhU9K/ScQsAP7hUibX39j7fakFPskvXusvfa0b4Q" crossorigin="anonymous"></script>
    <script src="https://maxcdn.bootstrapcdn.com/bootstrap/4.0.0/js/bootstrap.min.js" integrity="sha384-JZR6Spejh4U02d8jOt6vLEHfe/JQGiRRSQQxSfFWpi1MquVdAyjUar5+76PVCmYl" crossorigin="anonymous"></script>

    <script id="main" type="text/javascript">

    </script>

  </body>
</html>
```
